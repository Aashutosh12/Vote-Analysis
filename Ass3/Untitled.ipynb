```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.utils import resample
```

```python
df=pd.read_csv('diabetes.csv')
df
```

```python
df.info
```

```python
df.isnull().sum()
```

```python
df['Outcome'].value_counts()
```

```python
sns.countplot(data=df,x=df['Outcome'])
```

```python
one=df[df['Outcome']==1]
zero=df[df['Outcome']==0]
```

```python
one_upsample=resample(one,replace=True,n_samples=int(0.8*(len(zero))),random_state=42)
```

```python
new_df=pd.concat([zero,one_upsample])
```

```python
new_df
```

```python
new_df['Outcome'].value_counts()
```
